```python
#かい二乗分布
#自由度5,25,50に従うかい二乗分布が生成する乱数のヒストグラム
#[(5,'b'),(25,'g'),(50,'r')]というタプルの配列を作っている

import numpy as np 
import matplotlib.pyplot as plt
for df, c in zip([5,25,50],'bgr'):
    x = np.random.chisquare(df,1000)
    plt.hist(x,20,color=c)
plt.show()
    
```

```python
#t分布　自由度100,乱数1000
x = np.random.standard_t(100,1000)
plt.hist(x)
plt.grid(True)
```

```python
#F分布　自由度(10,30),(20,25) 乱数をそれぞれ1000
for df, c in zip([(10,30),(20,25)],'br'):
    x = np.random.f(df[0],df[1],1000)
    plt.hist(x,100,color=c)
    plt.grid(True)
```
