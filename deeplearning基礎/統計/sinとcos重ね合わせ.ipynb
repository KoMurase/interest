```python
import math
import matplotlib.pyplot as plt 
import numpy as np 
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
```

```python

x = np.linspace(-10,10,100)
plt.figure(figsize=(20,6))
plt.plot(x,sin_function(x))
plt.plot(x,cos_function(x))
plt.grid(True)
```
