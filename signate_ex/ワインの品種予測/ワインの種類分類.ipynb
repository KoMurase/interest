```python
import pandas as pd 
import numpy as np 
import numpy.random as random 
import scipy as sp

import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

%precision 3
```

```python
data = pd.read_csv('test.tsv',sep='\t')
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 

#æ¨™æº–åŒ–ã€€
sc = StandardScaler()
X_std = sc.fit_transform(data)

#ä¸»æˆåˆ†åˆ†æž
pca = PCA(n_components = 2)
pca.fit(X_std)
X_pca = pca.transform(X_std)

#è¡¨ç¤º
print(X_pca.shape)
print(pca.explained_variance_ratio_)
```

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(init='random',n_clusters=3)

kmeans.fit(X_pca)

labels = pd.Series(kmeans.labels_,name='cluster_number')+1

print(labels.value_counts(sort=False))
```

```python

y = kmeans.predict(X_pca)
y = pd.DataFrame(y)
```

```python
result = pd.concat([data['id'],y],axis=1)
```

```python
result.to_csv('./submit.csv',header=False,index=False)
```
